```python
def unify_victoria_secret(df):
    """
    We want that all brands that are related to Victoria's Secret
    have `victoria's secret` as their brand instead of what they
    currently have.
    """
    df = df.copy()
    new_string = "victoria's secret"
    # Write your code here
    variations_vs = ["Victorias-Secret", "Victoria's Secret", "victoria's secret Pink"]
    df["brand_name"] = df["brand_name"].replace(variations_vs, new_string, regex=True)
    return df
```

```python
df_unified = unify_victoria_secret(df)
```

## Cleaning up the price

In this section we are going to transform the `price` column into a float column in USD dolars.

For this, be careful of the different formats in the data.

For simplicity, you might assume that all the prices are in USD dolars, regarding of the symbol of the currency used.

```python
def clean_price(df):
    """
    In this function we will transform the
    `price` column into a column of floats.
    In case a product has more than one price,
    return the lowest one.
    """

    # Write your code here
    # Hacemos una copia del DataFrame para no modificar el original:
    df = df.copy()

    # Function to parse prices
    def parse_price(price_str):
        # Replace '$' and ' USD', and handle the string appropriately
        price_str = price_str.replace('$', '').replace(' USD', '').strip()
        # Split by space and convert to float
        prices = [float(p) for p in price_str.split() if p]  # Avoid empty strings
        return min(prices) if prices else np.nan  # Return min or NaN if empty

    # Apply the parsing function to the 'price' column
    df["price"] = df["price"].apply(parse_price)

    return df
```

```python
df_clean = clean_price(df_unified)
```

```python
answer_unified_price_sum = df_clean["price"].sum()
print(answer_unified_price_sum)
```

```python
def low_high_product_mean(df):
    """
    Finally, we will calculate `product_category` with the lowest and highest mean price
    for the brand Victoria's Secret.
    """

    # Write your code here
    # Filtrar el DataFrame para incluir solo la marca "victoria's secret":
    vs_df = df_unified[df_unified["brand_name"] == "victoria's secret"]

    # Calcular el precio promedio por categoría de producto:
    category_means = vs_df.groupby("product_category")["price"].mean()

    # Encontrar la categoría con el precio promedio más bajo y más alto:
    lowest_mean_category = category_means.idxmin()
    highest_mean_category = category_means.idxmax()

    # Obtener los valores de precio promedio más bajo y más alto
    lowest_mean = category_means[lowest_mean_category]
    highest_mean = category_means[highest_mean_category]

    return lowest_mean, highest_mean
```

```python
lowest_mean, highest_mean = low_high_product_mean(df_clean)
assert np.allclose(lowest_mean, 3.6203030303030)
print("highest_mean", highest_mean)
```
